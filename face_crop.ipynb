```python
from src import face_crop as fc

INPUT_DIR = './sample_input'  # 얼굴 크롭할 파일들 최상위 폴더
OUT_DIR = './outs'            # 크롭된 결과를 저장할 폴더
EXTS =['jpg', 'png', 'tiff', 'jpeg'] # 처리 대상이 되는 확장자. gif는 지원하지 않음
FACE_RESIZE = 1024
```

```python
# 처리할 파일 리스트 읽어 오기
files = fc.file_list(INPUT_DIR, EXTS)
print(f'처리할 파일 갯수: {len(files)} \n')


print('\n얼굴 탐지 및 피쳐 포인트 찾기')
face_info = fc.get_landmark(files)

print('\n크롭해서 저장하기')
SCALE = 0.9
croped = fc.save_croped_face(OUT_DIR, face_info, FACE_RESIZE, SCALE)
```

```python
# 샘플보기
import random
from PIL import Image
sample = random.choice(croped)
Image.open(sample)
```
